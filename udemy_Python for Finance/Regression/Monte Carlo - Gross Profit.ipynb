```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rev_m = 170
rev_stdev = 20
iterations = 1000
```

```python
rev = np.random.normal(rev_m, rev_stdev, iterations)
rev
```

```python
plt.figure(figsize=(15,6))
plt.plot(rev)
```

```python
COGS = - (rev * np.random.normal(0.6,0.1))

plt.figure(figsize=(15,6))
plt.plot(COGS)
```

```python
COGS.mean()
```

```python
COGS.std()
```

```python
gross_profit = rev + COGS
plt.figure(figsize=(15,6))
plt.plot(gross_profit)
```
